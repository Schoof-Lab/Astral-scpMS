### Exp5 - Cell type dependent coverage 

```python
#Importing packages 
import pandas as pd 
import numpy as np
import os 
import glob
import re
import seaborn as sns
import matplotlib.pyplot as plt
import umap
from scipy import stats
from scipy.stats import mstats
sns.set_style("ticks")

myColors = ["#7400b8","#6930c3","#5e60ce","#5390d9","#4ea8de","#48bfe3","#56cfe1","#64dfdf","#72efdd","#80ffdb"]
```

```python
#Data 
sc = pd.read_csv("data/Exp5/sc_stats.csv", sep = "\t",thousands='.')

#sc.to_csv("data/Exp5/output/sc_stats.tsv", sep = "\t", index = False)
```

```python
sc['CellType'] =  np.nan

sc['CellType'].loc[sc.iloc[:,0].str.contains(r'scHEK')] = 'HEK293'
sc['CellType'].loc[sc.iloc[:,0].str.contains(r'WT')] = 'u937'
sc['CellType'].loc[sc.iloc[:,0].str.contains(r'human')] = 'Human'
```

```python
fig, axes = plt.subplots(1,2, figsize = (10,5))

sns.barplot(data=sc, x = 'CellType', y = 'Peptides', palette = myColors, ax = axes[0])
sns.barplot(data=sc, x = 'CellType', y = 'Protein Groups', palette = myColors, ax = axes[1])
```

```python
sc.to_csv("data/Exp5/output/sc_stats.txt", sep = "\t", index = False)
```
